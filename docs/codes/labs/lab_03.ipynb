```python
import pandas as pd

# Cargar datos
df = pd.read_csv('https://raw.githubusercontent.com/fralfaro/MAT306/main/docs/labs/data/netflix_titles.csv')
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.head()
```

```python
df['year_added'] = df['date_added'].dt.year.astype('Int64')
df['month_added'] = df['date_added'].dt.month.astype('Int64')
df.head()
```

## Parte 2: Técnicas avanzadas de pandas

4. Utilizar `.loc` para seleccionar películas (`type == 'Movie'`) que fueron agregadas después del año 2018.

5. Utilizar `str.contains()` y `str.extract()`:

   * Filtrar títulos que contienen la palabra 'love' (sin distinguir mayúsculas/minúsculas).
   * Extraer la duración en minutos para las películas desde la columna `duration`.

6. Aplicar `explode()` sobre la columna `listed_in` para obtener una fila por cada género.

7. Obtener un top 10 de géneros más frecuentes utilizando `value_counts()`.

8. Aplicar `where()` y `mask()` para marcar las películas de más de 120 minutos como contenido largo en una nueva columna.

9. Utilizar `.loc` para filtrar películas que cumplen con:

   * Más de 100 minutos de duración.
   * Rating igual a `'R'`.
   * País igual a `'United States'`.

10. Utilizar `.style` para formatear visualmente el top 10 de películas más largas.

```python
df_new = df.loc[(df['year_added'] >= 2018) & (df['type']=='Movie')]
df.size,df_new.size
```

```python
df_new['year_added'].head()
```

```python
df_new = df.loc[df['title'].str.contains('Love', case=False)]
df.size,df_new.size
```

```python
df_new['title'].head(8)
```

```python
df_new = df.loc[df['type']=='Movie']
df_new['duration_minutes'] = df_new['duration'].str.extract(r'(\d+)').astype('Int64')
df_new['duration_minutes'].head()
```

```python
df_exploded = (
    df.assign(listed_in=df['listed_in'].str.split(r'\s*,\s*'))
      .explode('listed_in', ignore_index=True)
)
df.size,df_exploded.size
```

```python
df_exploded['listed_in'].head(8)
```

```python
df_exploded['listed_in'].value_counts(ascending=False).head(10)
```

```python
df_new['long_content'] = 'Normal'
df_new['long_content'] = df_new['long_content'].mask(df_new['duration_minutes'] > 120, 'Largo')
df_new['long_content'].head(8)
```

```python
df_new = df.loc[(df['type']=='Movie') & (df_new['duration_minutes'] > 100) & (df['rating']=='R') & (df['country']=='United States')]
df.size,df_new.size
```

```python
df_new.head()
```

```python
df_new = df.loc[df['type'] == 'Movie'].copy()
df_new['duration_minutes'] = df_new['duration'].str.extract(r'(\d+)').astype('Int64')

df_10_movies = df_new.sort_values('duration_minutes', ascending=False).head(10)
df_10_movies.style.background_gradient(cmap="Blues")
```
